```python
UnitB=pd.DataFrame(data1['Unit B'])
UnitB
```

```python
tStat,pValue =sp.stats.ttest_ind(UnitA,UnitB)
```

```python
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat))
```

```python
if pValue < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Wine.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import scale
norm=scale(dataset)
df = pd.DataFrame(norm,columns=dataset.columns)
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
```

```python
pca_values= pca.fit_transform(norm)
```

```python
pca_values
```

```python
explain_variance = pca.explained_variance_ratio_
explain_variance
```

```python
pca.components_
```

```python
explain_variance_1 = np.cumsum(np.round(explain_variance,decimals=4)*100)
explain_variance_1
```

```python
plt.plot(explain_variance_1,color='red')
```

```python
from sklearn.cluster import KMeans
```

```python
pca_df= pd.DataFrame(pca_values,columns=['pca1','pca2','pca2'])
pca_df
```

```python
#k-means
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i, init = 'k-means++' , random_state=0).fit(pca_df)
    wcss.append(kmeans.inertia_)
    print(kmeans)
    print(kmeans.inertia_)
    print(kmeans.labels_)
```

```python
import seaborn as sns
```

```python
sns.pointplot(list(range(1,10)),wcss)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendogram=sch.dendrogram(sch.linkage(pca_df,method='complete'))
plt.show()
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
y_hc=hc.fit_predict(pca_df)
```

```python
y_hc
```

```python
pca_df['label']=pd.Series(y_hc)
pca_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
book = pd.read_csv('book.csv')
book
```

#  apriori Algorithms 

```python
frequent_itemsets= apriori(book ,min_support=0.02, use_colnames=True,max_len=3)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
```

```python
plt.figure(figsize=(20 , 20 ))
plt.bar(x = list(range(1,11)),height = frequent_itemsets.support[1:11],color='rgmyk');
plt.xticks(list(range(1,11)),frequent_itemsets.itemsets[1:11])
plt.xlabel('item-sets')
plt.ylabel('support')
```

```python
sns.pairplot(book)
```

```python
len(rules)
```

```python
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules1
```

```python
rules[rules.lift>1]
```

```python
rules1[rules1.lift>2]
```

# Eclat Algorithms 

```python
transactions_1 = []
for sublist in book.values.tolist():
    clean_sublist = [item for item in sublist if item is not np.nan]
    transactions_1.append(clean_sublist)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
```

```python
te_ary = te.fit(book).transform(book)
te_ary
```

```python
df_x = pd.DataFrame(te_ary)
```

```python
df_x
```

```python
df_sets = apriori(df_x, min_support=0.002, use_colnames=True)
df_sets
```

```python
df_sets_1 = apriori(df_x, min_support=0.003, use_colnames=True)
df_sets_1
```

```python
df_rules = association_rules(df_sets,metric='support',min_threshold= 0.003,support_only=True) 
```

# pgrowth Algorithms 

```python
import pyfpgrowth
```

```python
transactions = []
for sublist in book.values.tolist():
    clean_sublist = [item for item in sublist if item is not np.nan]
    transactions.append(clean_sublist)
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions, 2)
patterns
```

```python
Fp_rules = pyfpgrowth.generate_association_rules(patterns, 0.5)
```
